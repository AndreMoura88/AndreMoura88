## Sobre o dataset
---
*About the dataset*

Informações sobre a base de dados<br>
*About this dataset* 
<br>

Para melhor acompanhamento das variáveis serão modificados seus nomes (o título terá uma preferência pelo idioma português) <br> 
*For better monitoring of the variables, their names will be modified (the title will have a preference for the Portuguese language)*
<br>

1. ```age``` <br>
Idade do paciente / *Age of the patient*
<br>

2. ```sex```<br>
Sexo do paciente/ *Sex of the patient*
* 0 = feminino/ *female*
* 1 = masculino/ *male*
<br>

3. ```tipo_de_dor```<br>
 Tipo de dor no peito/ *Chest Pain type*<br>
 Angina é o nome dado para a dor no peito causada pela diminuição do fluxo de sangue no coração.
 <br> *Angina is the name given to chest pain caused by decreased blood flow to the heart.*
* 0 = assintomático/ *asymptomatic* 
* 1 = angina típica/ *typical angina*
* 2 = angina atípica/ *atypical angina* 
* 3 = sem dor angina/ *non-anginal pain*
<br>

4. ```pres_art_repous```<br>
Pressão arterial em repouso / *resting blood pressure (in mm Hg)*
<br>

5. ```colest```<br>
Colesterol em mg/dl obtido via sensor de IMC / *cholestoral in mg/dl fetched via BMI sensor* 
<br>

6. ```acuc_sang_jejum```<br>
 Açúcar no sangue em jejum > 120 mg/dl / *fasting blood sugar > 120 mg/dl*
* 0 = Falso/ *False*
* 1 = Verdadeiro / *True*
<br>

7. ```eletrc_desc```<br>
Resultados eletrocardiográficos em repouso / *resting electrocardiographic results*
* 0 = hypertrophy,  mostrando hipertrofia ventricular esquerda provável ou definitiva pelos critérios de Estes / *showing probable or definite left ventricular hypertrophy by Estes' criteria*
* 1 = normal; 
* 2 = tendo anormalidade da onda ST-T (inversões da onda T e/ou elevação ou depressão ST > 0,05 mV) / *having ST-T wave abnormality, (T wave inversions and/or ST elevation or depression of > 0.05 mV)*
<br>

8. ```max_frq_card```<br>
Frequência cardíaca máxima alcançada / *maximum heart rate achieved* 
<br>

9. ```ang_indz_exerc```<br>
angina induzida por exercício / *exercise induced angina*
* 0 = Não/ *No*
* 1 = Sim/ *Yes*
<br>

10. ```ST_indz_exerc```<br>
Depressão de ST induzida por exercício em relação ao repouso / *ST depression induced by exercise relative to rest*
<br>

11. ```pico_ST_exerc```<br>
A inclinação do pico do segmento ST do exercício / *the slope of the peak exercise ST segment*
* 0 = ascendente/ *upsloping*
* 1 = plano /*flat*
* 2 = descendente/ *downsloping*
<br>

12. ```num_vasos```<br>
número de vasos principais  (0-3) / *number of major vessels*
<br>

13. ```talassemia```<br>
talassemia (uma forma de anemia crônica)/ *thalassemia*
* 1 = fixed defect
* 2 = normal
* 3 = reversable defect
<br>

14. ```output``` : 
* 0 = menos chance de um ataque do coração / *less chance of heart attack*<br>
* 1 = maior chance de um ataque do coração / *more chance of heart attack*

Verificando o tipo atual e conferindo a sugestão da biblioteca.

````python
report_on_dataframe(df_heart)
````

Como podemos observar o padrão do pandas não é a melhor escolha. Os ganhos no uso da memória pra esse dataset não é tão significante por se tratar de uma base de dados pequena. Mas por boas práticas adotaremos a sugestão do dtype_diet.  

*As we can see the pandas pattern is not the best choice. The gains in memory usage for this dataset are not that significant because it is a small database. But for good practices we will adopt dtype_diet suggestion*

````python
# salvando os tipos que queremos em um novo objeto
# saving the types we want in a new object
optmized_df = report_on_dataframe(df_heart, unit="MB")

# Utilizamos a otimização para transformar as colunas do 'df_heart' nas colunas de proposed_df
# We use the optimization to transform the 'df heart' columns into the proposed_df columns
heart_optimized = optimize_dtypes(df_heart, optmized_df)
````

Quando comparando os datasets, esse agora ocupa apenas 26% da memória que o primeiro, com as configurações padrão do pandas, usaria.

*When comparing the datasets, this one now takes up only 26% of the memory that the first one, with default pandas settings, would use.*

````python
pontos = [(20, 200), (30,190), (40, 180), (50, 170), (60, 160), (70, 150)]

grafico = sns.relplot(x='age', y='max_frq_card', data=df_heart)
grafico.set_ylabels("Frequência cardíaca Máxima")
grafico.set_xlabels("Age")

sns.regplot(data=df_heart, x='age', y='max_frq_card', ax=grafico.ax)

_ = sns.lineplot(x=[p[0] for p in pontos], y=[p[1] for p in pontos]
                 , color='green', ax=grafico.ax
                 , label='Média da frequência máx normal') 

plt.legend()
plt.show()
````

````python
# verificando os coeficientes de correlação e o valor de p
from scipy.stats import pearsonr

coef_corr, p_valor = pearsonr(df_heart['max_frq_card'], df_heart['age'])
print('Coeficiente de correlação:', round(coef_corr, 4), 'e o p-valor:', round(p_valor, 15))

print('\nÉ importante lembrar que o coeficiente de correlação por si só não\n\
indica causalidade, ou seja, não podemos afirmar que a idade é a causa da \n\
frequência cardíaca. Ele apenas indica que há uma relação linear entre \n\
as duas variáveis.')
````

````python
# correlação entre os valores forbes
coef_corr, p_valor = pearsonr(x=[p[0] for p in pontos], y=[p[1] for p in pontos])
print('Coeficiente de correlação:', round(coef_corr, 4), 'e o p-valor:', p_valor)
````
